```python
# Drop Unnamed column and check again
df.drop(columns=['Unnamed: 0'], inplace=True)

# Check if 'pdate' is in datetime format
df['pdate'] = pd.to_datetime(df['pdate'], errors='coerce')

# See unique values in 'label'
print(df['label'].value_counts())

# Drop ID column (optional for modeling)
df.drop(columns=['msisdn'], inplace=True)

# Optional: Print number of negative values per column
neg_counts = (df.select_dtypes(include='number') < 0).sum()
print("Negative value counts:\n", neg_counts[neg_counts > 0])
```

```python
# List of columns with negative values
cols_with_neg = ['aon', 'daily_decr30', 'daily_decr90', 'rental30', 'rental90',
                 'last_rech_date_ma', 'last_rech_date_da',
                 'medianmarechprebal30', 'medianmarechprebal90']

# Replace all negative values with 0
for col in cols_with_neg:
    df[col] = df[col].apply(lambda x: max(x, 0))

# Recheck if any negative values remain
print("Any negatives remaining?\n", (df[cols_with_neg] < 0).sum())
```
